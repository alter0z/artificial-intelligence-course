## sorting list practice

```python
lst = [1, 2, 4, -5, 7, 9, 3, 2]
newList = []

while lst:
  minVal = lst[0]
  for i in lst:
    if i < minVal:
      minVal = i
  
  newList.append(minVal)
  lst.remove(minVal)

print(newList)
```

## function

```python
def pow(x, y):
  return x**y

print(pow(2, 4) / 4)
```

```python
def fact(n):
  fact = 1
  for i in range(1, n + 1):
    fact *= i

  return fact

print(fact(5))
```

```python
def sortList(lst):
  newList = []

  while lst:
    minVal = lst[0]
    for i in lst:
      if i < minVal:
        minVal = i
    
    newList.append(minVal)
    lst.remove(minVal)
  return newList

print(sortList([1, 2, 4, -5, 7, 9, 3, 2]))
```

## numpy arange, random, reshape

### find det

```python
a = np.array([[7, -4, 2], [3, 1, -5], [2, 2, -5]])

def findDet(x):
  a, b, c, d, e, f, g, h, i = x[0, 0], x[0, 1], x[0, 2], x[1, 0], x[1, 1], x[1, 2], x[2, 0], x[2, 1], x[2, 2]
  return a*(e*i - f*h) - b*(d*i - f*g) + c*(d*h - e*g)

print(findDet(a))
```

### find inv

```python
b = np.array([[2, 6, 3], [4, -1, 3], [1, 3, 2]])

def transpose(x):
  a, b, c, d, e, f, g, h, i = x[0, 0], x[0, 1], x[0, 2], x[1, 0], x[1, 1], x[1, 2], x[2, 0], x[2, 1], x[2, 2]
  return np.array([[a, d, g], [b, e, h], [c, f, i]])

def cofactor(x):
  a, b, c, d, e, f, g, h, i = x[0, 0], x[0, 1], x[0, 2], x[1, 0], x[1, 1], x[1, 2], x[2, 0], x[2, 1], x[2, 2]
  return np.array([[(e*i - f*h), -(d*i - f*g), (d*h - e*g)], [-(b*i - c*h), (a*i - c*g), -(a*h - b*g)], [(b*f - c*e), -(a*f - c*d), (a*e - b*d)]])

def inv(b):
  detB = findDet(b)
  adjB = transpose(cofactor(b))

  return (1 / detB) * adjB

print(inv(b))
```

## linear regression

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
import numpy as np
```

```python
n = 30
x = np.arange(n)
y = 4 * x - 7 + (5*np.random.randn(n))
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression as lg

model = lg(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
yfit = model.predict(x[:, np.newaxis])
plt.scatter(x, y)
plt.plot(x, yfit, color='r')
```

## svm & random forest

```python
from sklearn.datasets._samples_generator import make_blobs as mb

x, y = mb(n_samples=500, centers=3, cluster_std=1.2)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='autumn')
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.svm import SVC

xTrain, xTest, yTrain, yTest = tts(x, y, test_size=0.33)

mdl = SVC(gamma='auto')
mdl.fit(xTrain, yTrain)

yPredict = mdl.predict(xTest)

acc = 100 * np.sum(yPredict == yTest) / yTest.size
print('accuracy @', acc, '%')
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc

mdl = rfc(n_estimators=20)
mdl.fit(xTrain, yTrain)

yPredict = mdl.predict(xTest)

acc = 100 * np.sum(yPredict == yTest) / yTest.size
print('accuracy @', acc, '%')
```
